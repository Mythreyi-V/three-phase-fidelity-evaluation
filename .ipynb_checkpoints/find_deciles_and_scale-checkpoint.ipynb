```python
def rule_breakdown(exp_list, feat_locs, feat_names, train_data):
    
    distribs = []
    
    for i in range(len(feat_locs)):
        feat_name = feat_names[i]
                
        min_vals = []
        max_vals = []

        for rule in exp_list:
            if feat_name in rule:

                loc = feat_locs[i]
    
                if rule.count("<") == 1:
                    min_vals.append(min([inst[loc] for inst in train_data]))
                    max_vals.append(float(rule.split("<")[-1].strip(" ").strip("=")))
                    
                elif rule.count("<") == 0 and rule.count(">") == 1:
                    min_vals.append(float(rule.split(">")[-1].strip(" ").strip("=")))
                    max_vals.append(max([inst[loc] for inst in train_data]))
                    
                elif rule.count("<") > 1:
                    min_vals.append(float(rule.split("<")[0].strip(" ").strip("=")))
                    max_vals.append(float(rule.split("<")[-1].strip(" ").strip("=")))
                    
        distribs.append((feat_locs[i], min(min_vals), max(max_vals)))
        
    return distribs
```

```python
def find_lime_distribution(instance, explainer, cls, quartiles, train_data):
    lime_exp = []
    for i in range(exp_iter):
        if classification==True:
            lime_exp.extend(explainer.explain_instance(instance, cls.predict_proba, 
                                                num_features=len(feat_list), labels=[0,1]).as_list())
        else:
            lime_exp.extend(explainer.explain_instance(instance, cls.predict, 
                                                    num_features=len(feat_list), labels=[0,1]).as_list())
            
    weights = [[] for each in feat_list]
    for exp in lime_exp:
        feat = exp[0].replace("= ",'')
        if '<' in feat:
            parts = feat.split('<')
        elif '>' in feat:
            parts = feat.split('>')
        
        for part in parts:
            if part.replace('.','').replace(' ','').isdigit()==False:
                feat_name = part.replace(' ','')
        n = feat_list.index(feat_name)
        weights[n].append(exp[1])
    
    weights = np.transpose(weights)
    avg_weight = np.average(np.array(weights), axis = 0)
    abs_weight = [abs(weight) for weight in avg_weight] 
        
    #For average explanation    
    bins = pd.cut(abs_weight, 10, retbins = True, duplicates = "drop")
    quartile_min = bins[1][quartiles]
    
    rel_features = [i for i in range(len(abs_weight)) if abs_weight[i] > quartile_min]    
    feat_names = [feat_list[i] for i in rel_features]
    
    rules = [exp[0] for exp in lime_exp]
    
    distribs = rule_breakdown(rules, rel_features, feat_names, train_data)
    
    return distribs


def find_shap_val_distribution(instance, explainer, quartiles, percentage, global_exp, train_data):
    full_exp = [shap_explainer(instance, check_additivity = False).values for i in range(exp_iter)]
    #print(full_exp)
    
    if classification==True:
        shap_exp = []
        for each in full_exp:
            single_exp = [feat[0] for feat in each]
            shap_exp.append(single_exp)
    else:
        shap_exp = full_exp

    avg_shap_exp = np.average(shap_exp, axis = 0)
    
    abs_exp = [abs(exp) for exp in avg_shap_exp]

    bins = pd.cut(abs_exp, 10, retbins = True, duplicates = "drop")
    quartile_min = bins[1][quartiles]

    rel_feats = [i for i in range(len(avg_shap_exp)) if abs_exp[i] > quartile_min]

    distribs = []
    
    for loc in rel_feats:
        orig_val = avg_shap_exp[loc]
        bin_size = abs(orig_val * percentage)
        min_val = orig_val-bin_size
        max_val = orig_val+bin_size

        feat_vals = []

        for i in range(len(global_exp)):
            if global_exp[i][loc] > min_val and global_exp[i][loc] < max_val:
                feat_vals.append(train_data[i][loc])
                
        max_feat = max(feat_vals)
        min_feat = min(feat_vals)

        distribs.append((loc, min_feat, max_feat))

    return distribs
```

```python
def find_feat_val_distribution(instance, explainer, explanation_method = "lime", cls = None, quartiles = 2, percentage = 0.1, global_exp = None, train_data = None):
    
    if quartiles == 1:
        bin_loc = -2
    elif quartiles == 2:
        bin_loc = -3
    elif quartiles == 3:
        bin_loc = -4
    elif quartiles == 4:
        bin_loc = -5
    elif quartiles == 5:
        bin_loc = -6
    elif quartiles == 6:
        bin_loc = -7
    elif quartiles == 7:
        bin_loc = -8
    elif quartiles == 8:
        bin_loc = -9
    else:
        bin_loc = -10
    
    if explanation_method == "shap":
        distribs = find_shap_val_distribution(instance, explainer, bin_loc, percentage, global_exp, train_data)
    
    elif explanation_method == "lime":
        distribs = find_lime_distribution(instance, explainer, cls, bin_loc, train_data)
        
    return distribs
```

```python
def get_tree_features(cls, instance):
    tree = cls.tree_
    lvl = 0
    left_child = tree.children_left[lvl]
    right_child = tree.children_right[lvl]

    feats = []
    val_lims = []
    
    while left_child != sklearn.tree._tree.TREE_LEAF and right_child != sklearn.tree._tree.TREE_LEAF:
        feature = tree.feature[lvl]
        feats.append(feature)
        val_lims.append((feature, tree.threshold[lvl]))

        if instance[feature] < tree.threshold[lvl]:
            lvl = left_child
        else:
            lvl = right_child
            
        left_child = tree.children_left[lvl]
        right_child = tree.children_right[lvl]
            
            
    feats = set(feats)
    
    return feats, val_lims
```

```python
# path to project folder
# please change to your own
PATH = os.getcwd()

dataset = "income"
cls_method = "decision_tree" 

classification = True

quartiles = [1, 2, 3, 4, 5, 6, 7, 8, 9]
percentage = np.arange(0.1, 2.1, 0.1)

random_state = 39
exp_iter = 10
random.seed(random_state)

save_to = "%s/%s/" % (PATH, dataset)
dataset_folder = "%s/datasets/" % (save_to)
final_folder = "%s/%s/" % (save_to, cls_method)

#Get datasets
X_train = pd.read_csv(dataset_folder+dataset+"_Xtrain.csv", index_col=False, sep = ";")
test_x = pd.read_csv(final_folder+"test_sample.csv", index_col=False, sep = ";").values

results = pd.read_csv(os.path.join(final_folder,"results.csv"), index_col=False)

feat_list = [each.replace(' ','_') for each in X_train.columns]

X = np.vstack((X_train.values, test_x))
```

```python
cls = joblib.load(save_to+cls_method+"/cls.joblib")
```

```python
if cls_method == "xgboost":
    shap_explainer = shap.Explainer(cls)
else:
    shap_explainer = shap.Explainer(cls, X, check_additivity = False)

all_shap_val = []
for i in range(exp_iter):
    exp = shap_explainer(X, check_additivity = False).values
    
    if classification==True:
        all_exp = []
        for each in exp:
            single_exp = np.array([feat[0] for feat in each])
            all_exp.append(single_exp)
    else:
        all_exp = exp
        
    all_shap_val.append(all_exp)

all_shap_exp = np.average(all_shap_val, axis = 0)

f1_scores = []

for quartile in quartiles:
    recall = []
    precision = []
    for instance in test_x:
        distribution = find_feat_val_distribution(instance, shap_explainer, explanation_method = "shap", global_exp = all_shap_exp, 
                                                  train_data = X, quartiles = quartile)

        true_features, val_lims = get_tree_features(cls, instance)
                
        exp_features = [feat[0] for feat in distribution]
        
        recall.append(len(true_features.intersection(exp_features))/len(true_features))
        precision.append(len(true_features.intersection(exp_features))/len(exp_features))

    f1 = 2*((np.mean(recall)*np.mean(precision))/(np.mean(recall)+np.mean(precision)))
    f1_scores.append(f1)
    
    print(quartile, "deciles: Recall = ", np.mean(recall), "Precision = ", np.mean(precision), "F1 = ", f1)
    print("-------------------------------------------------------------------")

loc = f1_scores.index(max(f1_scores))
opt_quartile = quartiles[loc]

print("Best decile: %s with average F1 of %s" % (opt_quartile, max(f1_scores)))
```

```python
all_shap_val = []
for i in range(exp_iter):
    exp = shap_explainer(X, check_additivity = False).values
    
    if classification==True:
        all_exp = []
        for each in exp:
            single_exp = np.array([feat[0] for feat in each])
            all_exp.append(single_exp)
    else:
        all_exp = exp
        
    all_shap_val.append(all_exp)

all_shap_exp = np.average(all_shap_val, axis = 0)

quartile = opt_quartile

avg_diff = []

for percent in percentage:
    differences = []
    for instance in test_x:
        distribution = find_feat_val_distribution(instance, shap_explainer, explanation_method = "shap", global_exp = all_shap_exp, 
                                                  train_data = X, quartiles = quartile,  percentage = percent)

        true_features, val_lims = get_tree_features(cls, instance)

        for i in range(len(distribution)):
            each = list(distribution[i])
            feat_name = each[0]
            min_val = each[1]
            max_val = each[2]
            
            thresholds = []
            for threshold in val_lims:
                if feat_name in threshold:
                    true_threshold = threshold[1]
                    thresholds.append(true_threshold)
            
                    if len(thresholds) == 1:
                        true_threshold = thresholds[0]
                        if instance[feat_name] > true_threshold:
                            diff = (min_val - true_threshold)/true_threshold
                            differences.append(abs(diff))
                        else:
                            diff = (true_threshold - max_val)/true_threshold
                            differences.append(abs(diff))
                    else:
                        min_threshold = min(thresholds)
                        max_threshold = max(thresholds)

                        if instance[feat_name] < min_threshold:
                            diff = (min_threshold - max_val)/min_threshold
                            differences.append(abs(diff))
                        elif instance[feat_name] > min_threshold and instance[feat_name] < max_threshold:
                            diff1 = (min_val - min_threshold)/min_threshold
                            diff2 = (max_threshold - max_val)/max_threshold
                            differences.append(abs(diff1))
                            differences.append(abs(diff2))
                        elif instance[feat_name] > max_threshold:
                            diff = (min_val - max_threshold)/max_threshold
                            differences.append(abs(diff))
                    
                
    print("Bin size: %s, Average distance to/from thresold: %s" % (percent, np.mean(differences)))
    avg_diff.append(np.mean(differences))
    print("-------------------------------------------------------------------")
    
loc = avg_diff.index(min(avg_diff))
opt_percentage = percentage[loc]

print("Best percentage: %s with average difference of %s" % (opt_percentage, min(avg_diff)))
```
